```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# 상품 정보를 수집하는 사용자 정의 함수
def scrape_products(driver, target_count=100):
    # 빈 리스트와 중복 확인을 위한 set
    box = []
    collected_items = set()

    # 상품 수집 반복
    while len(box) < target_count:
        # 페이지를 아래로 스크롤
        body = driver.find_element(By.CSS_SELECTOR, 'body')
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)  # 페이지 로드 대기

        # BeautifulSoup으로 HTML 파싱
        html = BeautifulSoup(driver.page_source, 'html.parser')

        # 상품 정보 추출
        products = html.find_all('div', class_='_2a0vHVAYuk')  # 해당 클래스명을 가진 상품 영역

        for product in products:
            title_tag = product.find('strong', class_='_1lxNNH5gfD')
            price_tag = product.find('span', class_='_3MzrqSUlwR')
            rating_tag = product.find('span', class_='_29gNnX1yLx')

            if title_tag and price_tag and rating_tag:
                # 상품 제목, 가격, 평점 추출
                title = title_tag.get_text(strip=True)
                price = price_tag.get_text(strip=True).replace('원', '')
                rating = rating_tag.get_text(strip=True).replace('평점', '')

                # 중복된 상품인지 확인 후 저장
                if title not in collected_items:
                    box.append({
                        'title': title,
                        'price': price,
                        'rating': rating
                    })
                    collected_items.add(title)

            # 100개 이상 수집되면 중단
            if len(box) >= target_count:
                break

    # DataFrame으로 변환하여 반환
    return pd.DataFrame(box)


```

```python
# WebDriver 설정 (Chrome 예시)
# driver = webdriver.Chrome()

# 사이트로 이동
# driver.get("https://shopping.naver.com/fresh/directfarm/home")  # 크롤링할 사이트의 URL 입력

# 상품 정보 수집 함수 호출
df = scrape_products(driver, target_count=100) # 상품 100개

# 데이터 프레임을 Excel 파일로 저장
df.to_excel('scraped_products(naver).xlsx', index=False)

df
```

```python
from tqdm import tqdm
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd

total = []

for i in range(2):  # sosok=0 (KOSPI), sosok=1 (KOSDAQ)
    if i == 0:
        entity = 'KOSPI'
    else:
        entity = 'KOSDAQ'
        
    print(f"Fetching data for {entity}...")

    for n in tqdm(range(1, 48)):  # 최대 47페이지까지 시도
        dic = {"menu": 'market_sum',
               "returnUrl": f"http://finance.naver.com/sise/sise_market_sum.naver?sosok={i}&page={n}",
               "fieldIds": fields}

        url = requests.post("https://finance.naver.com/sise/field_submit.naver", data=dic)
        html = BeautifulSoup(url.text, 'html.parser')

        table = html.find("table", class_='type_2')

        if table is None:  # 페이지가 아예 없으면 다음 루프로
            print(f"No table found for {entity} at page {n}. Moving to next.")
            continue

        table = pd.read_html(str(table))[0]

        # 데이터가 비어있으면 다음 루프로
        if table.empty or len(table) == 0:
            print(f"No data for {entity} at page {n}. Moving to next.")
            continue

        # Unnamed 열 제거
        table = table.loc[:, ~table.columns.str.contains('^Unnamed')]

        # 종목명 앞에 KOSPI/KOSDAQ 추가
        table['종목'] = f"{entity} "

        total.append(table)
        time.sleep(2)  # 서버에 부담을 주지 않기 위해 잠시 대기

print("All pages processed. No more data to fetch.")
```

```python
table = pd.concat(total)
table
```

```python
# '토론실' 열 삭제
stock = table.drop(columns=['토론실'])

# '종목명'이 결측치인 행 삭제
stock = stock.dropna(subset=['종목명'])

# 결과 확인
stock
```

```python
stock.to_excel('stock_PBR.xlsx')
```

```python
# 'PER' 열이 0 이상 1 미만인 종목만 필터링
filtered_stock = stock[(stock['PBR'] >= 0) & (stock['PBR'] < 1)]

# PER 오름차순으로 정렬
sorted_stock = filtered_stock.sort_values(by='PBR', ascending=True)

# 결과 출력
sorted_stock
```

```python
from datetime import datetime



# 파일 이름에 현재 날짜 포함하여 저장
filename = f"{str(datetime.today()).split()[0]} 국내주식 PBR 1 미만 저평가 종목 리스트.txt"
f = open(filename, 'wt')

# 각 종목에 대해 저장할 내용을 생성하고 파일에 기록
for idx, row in sorted_stock.iterrows():
    line = f"{row['종목명']} ({row['종목']}) PER : {row['PBR']}\n"
    f.write(line)

f.close()  # 파일 저장하고 닫기

print(f"저평가 종목 리스트가 {filename} 파일로 저장되었습니다.")
```
